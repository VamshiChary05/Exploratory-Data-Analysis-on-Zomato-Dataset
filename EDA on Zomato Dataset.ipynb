```python
# Dataset 2
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
# we have a common column in both the dataset which is 'Country Code'
```

```python
# Let's Combine the both datasets using common feature.
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how ='left')
# I am doing left merge because more data is present in dataset 1(more Information).
```

```python
final_df.head()
```

```python
# lets find datatypes
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
# with the help of value_counts we can find how many records contain which country.
```

```python
# As per this dataset contain more delivery information on India then USA then UK
```

```python
Country_names = final_df.Country.value_counts().index
```

```python
Country_names
```

```python
Values = final_df.Country.value_counts().values
```

```python
Values
```

```python
## Creating PIE CHART - Top 5 Countries that uses Zomato Services
plt.pie(Values[:5],labels = Country_names[:5],autopct='%1.2f%%')
plt.show()
```

```python
# 93.17% of transactions are from India.
# 4.67% of transactions are from United States. then follows United Kingdom.
```

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
#Observations
# When Ratings is b/w 4.5 to 4.9 is Excellent.
# When Ratings is b/w 4.0 to 3.4 is Very Good.
# When Ratings is b/w 3.5 to 3.9 is Good.
# When Ratings is b/w 3.0 to 3.4 is Average.
# When Ratings is b/w 2.5 to 2.9 is Average.
# When Ratings is b/w 2.0 to 2.4 is Poor.
# When Rating is 0.0 that means no ratings given by customer.
```

```python
sns.barplot(x='Aggregate rating',y= 'Rating Count',data= ratings)
plt.show()
```

```python
#Obseravtions:
# 1. Not Rated Count is very High
# 2. Maximium number of ratings are from 2.5 to 4.4
```

```python
sns.countplot(x='Rating color',data= ratings,palette=['blue','red','orange','yellow','green','green'])
plt.show()
```
